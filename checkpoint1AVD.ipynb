## Studi Kasus




Seorang konten kreator ingin memahami bagaimana aktivitas posting di media sosial mempengaruhi engagement yang diperoleh, khususnya jumlah likes. Dengan semakin banyaknya pengguna media sosial, penting untuk mengetahui pola aktivitas yang efektif dalam meningkatkan interaksi dengan audiens.

Melalui dataset Social Media Usage Dataset, analisis dilakukan untuk melihat hubungan antara frekuensi posting per hari dengan jumlah likes yang diterima pengguna. Hasil analisis ini diharapkan dapat memberikan gambaran mengenai pola aktivitas media sosial yang berkaitan dengan engagement.

### Business Objective

Dataset Social Media Usage Dataset berisi data aktivitas pengguna media sosial pada beberapa platform, seperti waktu penggunaan harian, jumlah posting, jumlah likes, dan jumlah followers yang diperoleh per hari.

Dari sisi bisnis, engagement seperti jumlah likes menjadi indikator penting dalam mengukur performa konten di media sosial.

Tujuan analisis bisnis:
- Memahami hubungan antara frekuensi posting dan jumlah likes
- Mengidentifikasi pola aktivitas pengguna media sosial
- Mengetahui faktor aktivitas yang berkaitan dengan engagement pengguna

### Assess Situation

Dataset ini berisi data penggunaan media sosial dari 1.000 pengguna pada tujuh platform populer seperti Facebook, Instagram, Twitter, Snapchat, TikTok, LinkedIn, dan Pinterest.

Dataset mencatat beberapa aktivitas pengguna media sosial, seperti:
- Daily_Minutes_Spent: waktu penggunaan aplikasi per hari
- Posts_Per_Day: jumlah posting per hari
- Likes_Per_Day: jumlah likes yang diterima per hari
- Follows_Per_Day: jumlah followers baru per hari

Dataset ini dapat digunakan untuk melihat pola aktivitas pengguna media sosial dan tingkat engagement yang diperoleh.

Beberapa tantangan dalam dataset ini antara lain:
- Data kemungkinan bersifat simulasi sehingga mungkin tidak sepenuhnya menggambarkan kondisi nyata.
- Dataset hanya berisi aktivitas dasar pengguna media sosial.
- Tidak terdapat informasi tambahan seperti jenis konten atau data pengguna.
- Perlu dilakukan pengecekan data sebelum analisis, seperti memastikan tidak ada data kosong atau duplikat.
