### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Terlihat pada dataset ini tidak terdapat outliers pada seluruh kolom numerik, karena persentase outliers adalah 0%. Maka hal ini tidak perlu ditangani.

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas:
Membandingkan rata-rata jumlah likes berdasarkan kategori frekuensi posting (rendah, sedang, tinggi).

Tujuan:
Mengetahui apakah kategori dengan frekuensi posting lebih tinggi memiliki rata-rata likes yang lebih besar dibandingkan kategori lainnya.

Visualisasi:
Bar Chart

```python
df['Kategori_Posting'] = pd.qcut(
    df['Posts_Per_Day'],
    q=3,
    labels=['Rendah', 'Sedang', 'Tinggi']
)

# Menghitung rata-rata likes berdasarkan kategori posting
perbandingan = df.groupby(
    'Kategori_Posting',
    observed=False
)['Likes_Per_Day'].mean()

# Visualisasi bar chart
plt.figure(figsize=(8,6))
sns.barplot(
    x=perbandingan.index,
    y=perbandingan.values
)

plt.title('Rata-rata Likes Berdasarkan Kategori Frekuensi Posting')
plt.xlabel('Kategori Frekuensi Posting')
plt.ylabel('Rata-rata Likes')
plt.show()
```

Berdasarkan grafik batang di atas, terlihat bahwa kategori Tinggi memiliki rata-rata jumlah likes paling besar dibandingkan kategori lainnya. Nilainya berada sedikit di atas kategori Sedang dan Rendah, meskipun selisihnya tidak terlalu jauh.

Di posisi berikutnya terdapat kategori Sedang, yang menunjukkan rata-rata likes yang hampir setara dengan kategori Tinggi. Perbedaannya sangat tipis dan tidak terlalu signifikan secara visual.

Sementara itu, kategori Rendah memiliki rata-rata likes paling kecil dibandingkan dua kategori lainnya. Namun, perbedaannya juga relatif kecil dan tidak menunjukkan gap yang mencolok.

Kesimpulan : Distribusi rata-rata likes antar kategori frekuensi posting relatif tidak terlalu jauh perbedaannya. Meskipun kategori dengan frekuensi posting tinggi cenderung memiliki rata-rata likes lebih besar, peningkatannya tidak signifikan. Hal ini mengindikasikan bahwa frekuensi posting bukan faktor utama yang secara kuat memengaruhi jumlah likes dalam dataset ini.

## **Composition/Komposisi**

Aktivitas:
Melihat proporsi jumlah likes berdasarkan kategori frekuensi posting pengguna di media sosial.

Tujuan:
Mengetahui seberapa besar kontribusi setiap kategori frekuensi posting terhadap total jumlah likes, sehingga dapat menilai apakah frekuensi posting yang lebih tinggi cenderung mendapatkan lebih banyak likes.

Visualisasi:

Berdasarkan visualisasi pie chart, terlihat bahwa:

* Disusul oleh kategori 19 posting per hari sebesar 33,2%.

* Kemudian kategori 16 posting per hari sebesar 33,8%, yang menjadi kontribusi terbesar.

Perbedaan persentase di antara ketiga kategori ini sangat kecil dan relatif seimbang, masing-masing berada di kisaran 33%.

Kesimpulan :

Jumlah posting sebanyak 16–19 kali per hari sama-sama memberikan kontribusi besar terhadap total perolehan likes. Namun, tidak terdapat perbedaan yang signifikan antar ketiganya.

Hal ini menunjukkan bahwa pada tingkat frekuensi tinggi, peningkatan jumlah posting tidak selalu menghasilkan lonjakan likes yang drastis, karena kontribusinya cenderung merata.

Dengan kata lain, dalam rentang frekuensi tersebut, performa likes relatif stabil dan tidak menunjukkan dominasi yang kuat dari satu kategori tertentu.

## **Relationship/Hubungan**

Aktivitas:
Menganalisis korelasi antara frekuensi posting (Posts_Per_Day) dengan jumlah likes per hari (Likes_Per_Day).

Tujuan:
Mengetahui apakah peningkatan frekuensi posting berbanding lurus dengan peningkatan jumlah likes, serta mengidentifikasi kemungkinan adanya aktivitas posting yang tinggi namun tidak menghasilkan engagement (likes) yang signifikan.

Visualisasi:
Heatmap (Grafik Korelasi)

```python
plt.figure(figsize=(8, 6))

sns.heatmap(
    df[['Posts_Per_Day', 'Likes_Per_Day']].corr(),
    annot=True,
    cmap='viridis',
    fmt='.2f'
)

plt.title('Korelasi antara Frekuensi Posting dan Jumlah Likes')
plt.show()
```

Berdasarkan heatmap korelasi, nilai korelasi antara frekuensi posting (posts_per_day) dan jumlah likes (likes_per_day) adalah sebesar 0.02.

Nilai ini menunjukkan bahwa hubungan antara kedua variabel tersebut sangat lemah dan hampir tidak ada korelasi. Angka 0.02 mendekati nol, yang berarti peningkatan frekuensi posting tidak secara signifikan diikuti oleh peningkatan jumlah likes.

Interpretasi :

Hasil ini mengindikasikan bahwa dalam dataset yang digunakan, banyaknya posting yang dilakukan pengguna tidak secara langsung memengaruhi jumlah likes yang diperoleh.

Dengan kata lain:

Pengguna yang sering posting belum tentu mendapatkan likes lebih banyak.

Begitu juga pengguna yang jarang posting tidak selalu mendapatkan likes yang sedikit.

Kemungkinan terdapat faktor lain yang lebih berpengaruh terhadap jumlah likes, seperti kualitas konten, waktu posting, jenis platform, atau jumlah followers.

Kesimpulan :

Berdasarkan analisis korelasi, frekuensi posting tidak memiliki pengaruh yang signifikan terhadap jumlah likes dalam dataset ini. Hal ini menunjukkan bahwa peningkatan jumlah posting saja tidak cukup untuk meningkatkan engagement berupa likes.

Walaupun Snapchat memiliki persentase paling tinggi, sebenarnya perbedaan antara ketiga platform tersebut tidak terlalu jauh. Hal ini menunjukkan bahwa interaksi pengguna berupa likes cukup tersebar di beberapa platform media sosial dan tidak hanya terpusat pada satu platform saja.
